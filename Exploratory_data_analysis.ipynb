```python
#we are keeping the older releases and higher number of votes

dataframe = dataframe.sort_values(by='vote_count', ascending=False)
dataframe.drop_duplicates(subset=['title'], keep='first', inplace=True)
dataframe.duplicated(subset=['title'], keep=False).sum()
```

```python
dataframe=dataframe.sort_values(by='id').reset_index(drop=True)
dataframe.head()
```

```python
dataframe.info()
```

```python
dataframe['release_date']=pd.to_datetime(dataframe['release_date'])
dataframe.dtypes
```

**Feature Engineering**

```python
#creating two columns from "release_date" to use it for insights

dataframe['decade'] = (dataframe['release_date'].dt.year//10*10).astype(str)+'s'
dataframe.head()
```

```python
dataframe['release_year']=dataframe['release_date'].dt.year
dataframe.head()
```

# Visualisation

```python
#1. distribution of popularity
plt.figure(figsize=(15,5))
sns.histplot(x='popularity', data=dataframe, bins=50)
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.title('Distribution of Popularity')
plt.show()
```

```python
#2. number of movies per decades
plt.figure(figsize=(10,5))
sns.barplot(x=dataframe['decade'].value_counts().index, y=dataframe['decade'].value_counts().values)
plt.xlabel('decade')
plt.ylabel('number of movies')
plt.title('number of movies per decade')
plt.xticks(rotation=45)
plt.show()
```

```python
#3. relationship between decades and rating
plt.figure(figsize=(15,5))
sns.boxplot(x='decade', y='vote_average', data=dataframe)
plt.xlabel('decade')
plt.ylabel('average vote')
plt.title('relationship between decade and average number of votes')
plt.show()
```

```python
#4. number of movies over the years
counts_year = dataframe['release_date'].dt.year.value_counts().sort_index()
plt.figure(figsize=(15,5))
sns.lineplot(x=counts_year.index, y=counts_year.values, marker='o')
plt.xlabel('year')
plt.ylabel('number of movies')
plt.title('trend of movies released over the years')
plt.xticks(counts_year.index[::5], rotation=90)
plt.grid(True)
plt.show()
```

```python
#5. visualising top 5 movies different columns
mov_top = dataframe.nlargest(5, 'vote_count')
plt.figure(figsize=(8,5))
sns.barplot(data=mov_top, x='vote_count', y='title', palette='magma')
plt.title('top 5 movies by vote count')
plt.xlabel('vote count')
plt.ylabel('movie title')
plt.show()
```

```python
#relationship between "popularity" and "vote_average"
plt.figure(figsize=(8,5))
sns.scatterplot(data=dataframe, x='popularity', y='vote_average', palette='Blues')
plt.title('Top 5 most popular movies')
plt.xlabel('popularity')
plt.ylabel('Movie title')
plt.show()
```

```python
#most movies have low popularity and votes between 6.0 and 8.5, but there are also
#movies with high popularity as well as high number of votes, which inidcates little to no correlation
```

```python
#correlation between popularity and vote count
plt.figure(figsize=(8,5))
corr = dataframe[['popularity', 'vote_average', 'vote_count']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('feature correlation')
plt.show()
```

```python
#the above heatmap shows weak correlation between movie popularity and votes
```

```python
#movie rating distribution
plt.figure(figsize=(15,5))
sns.histplot(dataframe['vote_average'], bins=20, kde=True, color='blue')
plt.title('movie rating distribution')
plt.xlabel('rating')
plt.ylabel('count')
plt.show()
```

```python
#average popularity over the years
year_pop = dataframe.groupby('release_year')['popularity'].mean()
plt.figure(figsize=(12,5))
sns.lineplot(x=year_pop.index, y=year_pop.values, marker='o', color='red')
plt.xlabel('release year')
plt.ylabel('average popularity')
plt.title('average popularity of movies over time')
plt.grid(True)
plt.show()
```
