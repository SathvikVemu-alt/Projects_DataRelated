```python
#5. Figuring out the most frequent directors
plt.figure(figsize=(12,5))
spnetpd['director'].dropna().value_counts().nlargest(10).plot(kind='bar', color='purple')
plt.title('Top 10 most featured directors')
plt.xlabel('director')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```

```python
from wordcloud import WordCloud
#6. Most frequent actors
txt = ' '.join(spnetpd['cast'].dropna())
cloud = WordCloud(width=800, height=400, background_color='black').generate(txt)
plt.figure(figsize=(10,5))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.title('most frequent actors')
plt.show()
```

```python
#7. rating distribution
plt.figure(figsize=(10,5))
sns.countplot(data=spnetpd, y='rating', order=spnetpd['rating'].
              value_counts().index, palette='coolwarm')
plt.title('content rating distribution')
plt.xlabel('count')
plt.ylabel('rating')
plt.show()
```

```python
#8. duration analysis
mov_dur =spnetpd[spnetpd['type'] == 'Movie'].copy()
mov_dur['duration'] = mov_dur['duration'].str.replace(' min', '').astype(float)
plt.figure(figsize=(10,5))
sns.histplot(mov_dur['duration'].dropna(), bins=30, color='red', kde=True)
plt.title('distribution of movie durations')
plt.xlabel('duration (minutes)')
plt.ylabel('count')
plt.show()
```

```python
#9. Most popular genres
plt.figure(figsize=(12,5))
genres = spnetpd['listed_in'].str.split(', ', expand=True).stack()
genres.value_counts().nlargest(10).plot(kind='bar', color='orange')
plt.title('top 10 most common genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#10. description text analysis (word cloud)
desc_text = ' '.join(spnetpd['description'].dropna())
wordcloud_desc = WordCloud(width=800, height=400, background_color='black').generate(desc_text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud_desc, interpolation='bilinear')
plt.axis('off')
plt.title('common words in descriptions')
plt.show()
```

# Predictive Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix

net = spnetpd.dropna(subset=['type'])
net = net.fillna('unknown')

colcat = ['director', 'cast', 'country', 'rating', 'listed_in']
enc = LabelEncoder()

for col in colcat:
  net[col] = enc.fit_transform(net[col].astype(str))
```

```python
#converting date_added to year
net['date_added'] = pd.to_datetime(net['date_added'], errors='coerce')
net['year_added'] = net['date_added'].dt.year.fillna(0).astype(int)
```

```python
#features and target
X = net[['director', 'cast', 'country', 'rating', 'listed_in', 'year_added', 'release_year']]
Y = enc.fit_transform(net['type'])
```

```python
#data split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
# defining models
models = {
    'logreg': LogisticRegression(max_iter=1000),
    'rf': RandomForestClassifier(n_estimators=100),
    'gb': GradientBoostingClassifier(),
    'svm': SVC()
}
```

```python
#TRAINING AND MODEL evaluation
for name, model in models.items():
  model.fit(Xtrain, Ytrain)
  Ypred = model.predict(Xtest)
  acc = accuracy_score(Ytest, Ypred) * 100
  print(f'{name} accuracy: {acc:.2f}%')
```

```python
#confusion matrix
cm = confusion_matrix(Ytest, Ypred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Movie', 'TV Show'],
            yticklabels=['Movie', 'TV Show'])
plt.title(f'confusion matrix -{name}')
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.show()
```
